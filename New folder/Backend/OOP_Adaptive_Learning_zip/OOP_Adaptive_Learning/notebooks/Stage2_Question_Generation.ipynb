```python
topics = [
    "Principles of OOP",
    "Classes & Objects",
    "Inheritance & Polymorphism",
    "OOP Core Concepts",
    "Collections & Generics"
]

difficulty_levels = {
    "Level 1": "Beginner",
    "Level 2": "Intermediate",
    "Level 3": "Advanced"
}
```

```python
def generate_prompt(topic, difficulty, question_type="MCQ"):
    if question_type == "MCQ":
        return f"""
        Generate a {difficulty} level multiple-choice question on Java for the topic: {topic}.
        Requirements:
        - 1 correct answer and 3 wrong options
        - Indicate the correct answer
        Example format:
        Q: <Question>
        A) ...
        B) ...
        C) ...
        D) ...
        Correct Answer: B
        """
    elif question_type == "Coding":
        return f"""
        Generate a {difficulty} level Java coding problem on the topic: {topic}.
        Include:
        - Problem statement
        - Expected output example
        """
```

```python
def generate_question(prompt):
    try:
        response = openai.ChatCompletion.create(
            model="gpt-3.5-turbo",  # You can upgrade to GPT-4 later
            messages=[{"role": "system", "content": "You are a Java programming expert and question setter."},
                      {"role": "user", "content": prompt}],
            temperature=0.7
        )
        return response['choices'][0]['message']['content'].strip()
    except Exception as e:
        return f"Error: {e}"
```

```python
# Example: On-demand generation for one topic
topic = "Inheritance & Polymorphism"
difficulty = difficulty_levels["Level 1"]

# MCQ
prompt_mcq = generate_prompt(topic, difficulty, "MCQ")
mcq_question = generate_question(prompt_mcq)

# Coding Question
prompt_code = generate_prompt(topic, difficulty, "Coding")
coding_question = generate_question(prompt_code)

print("MCQ Question:\n", mcq_question)
print("\nCoding Question:\n", coding_question)
```
