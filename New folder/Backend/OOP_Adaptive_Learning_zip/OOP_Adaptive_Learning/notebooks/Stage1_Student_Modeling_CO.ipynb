```python
import pandas as pd
import json
from utils.co_topic_mapping import CO_TOPIC_MAP

# Load CO-based dataset
file_path = r"E:\Capstone\OOP_Adaptive_Learning\data\student_co_marks.xlsx"
df = pd.read_excel(file_path)

# Thresholds for classification
weak_threshold = 10
moderate_threshold = 15

student_profiles = []

for _, row in df.iterrows():
    student_id = row["Student_ID"]
    weak_topics, moderate_topics, strong_topics = [], [], []

    for co, topics in CO_TOPIC_MAP.items():
        score = row[co]
        if score < weak_threshold:
            weak_topics.extend(topics)
        elif score < moderate_threshold:
            moderate_topics.extend(topics)
        else:
            strong_topics.extend(topics)

    student_profiles.append({
        "student_id": int(student_id),
        "weak_topics": weak_topics,
        "moderate_topics": moderate_topics,
        "strong_topics": strong_topics
    })

# Save profiles
with open("../data/student_profiles.json", "w") as f:
    json.dump(student_profiles, f, indent=4)

print("✅ Profiles updated based on CO mapping!")
```
