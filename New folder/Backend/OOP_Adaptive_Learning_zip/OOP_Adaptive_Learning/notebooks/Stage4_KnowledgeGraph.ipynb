```python
import networkx as nx
import matplotlib.pyplot as plt
import textwrap

# Knowledge Graph setup
G = nx.DiGraph()
topics = {
    "Principles of OOP": "CO1",
    "Classes & Objects": "CO2",
    "Inheritance & Polymorphism": "CO2",
    "Collections & Generics": "CO3"
}
for topic, co in topics.items():
    G.add_node(topic, co=co)

G.add_edge("Principles of OOP", "Classes & Objects")
G.add_edge("Classes & Objects", "Inheritance & Polymorphism")
G.add_edge("Inheritance & Polymorphism", "Collections & Generics")

# Layout with good spacing
pos = nx.spring_layout(G, seed=42, k=2)  # increase spacing

plt.figure(figsize=(14, 8))

# Draw nodes and edges
nx.draw_networkx_nodes(G, pos, node_size=4000, node_color="skyblue", edgecolors="black")
nx.draw_networkx_edges(G, pos, arrowstyle="->", arrowsize=30, edge_color="gray", width=2)

# Wrap labels for readability
node_labels = {node: f"{textwrap.fill(node, width=18)}\n({data['co']})" for node, data in G.nodes(data=True)}
nx.draw_networkx_labels(G, pos, labels=node_labels, font_size=13, font_weight="bold")

# Edge labels
edge_labels = {(u, v): "prerequisite" for u, v in G.edges()}
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=12)

plt.title("Knowledge Graph for OOP Topics", fontsize=18, fontweight="bold")
plt.axis("off")
plt.show()
```
